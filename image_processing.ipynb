```
import cv2

# Load the image
img = cv2.imread('fish.png')

# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(img_rgb)
plt.show()
```

In this step, we will determine the size of the image 'fish.png'.

```
# Get the size of the image
height, width, _ = img.shape

print(f'The image size is {width}x{height}')
```

In this step, we will resize the image 'fish.png' by 50%.

```
# Resize the image
resized_img = cv2.resize(img, (width // 2, height // 2))

# Convert the image from BGR to RGB
resized_img_rgb = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)

# Display the resized image
plt.imshow(resized_img_rgb)
plt.show()
```

In this step, we will convert the resized image to grayscale.

```
# Convert the image to grayscale
gray_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_img, cmap='gray')
plt.show()
```

In this step, we will perform an operation on the grayscale image where each pixel value is replaced by (255 - pixel value). This operation is also known as image negation.

```
# Perform the image negation operation
neg_img = 255 - gray_img

# Display the negated image
plt.imshow(neg_img, cmap='gray')
plt.show()
```

In this step, we will display the grayscale image and the inverted image side by side for comparison.

```
# Display the grayscale image and the inverted image side by side
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(gray_img, cmap='gray')
axs[0].set_title('Grayscale Image')
axs[0].axis('off')

axs[1].imshow(neg_img, cmap='gray')
axs[1].set_title('Inverted Image')
axs[1].axis('off')

plt.show()
```

In this step, we will apply edge detection to the grayscale image using the Canny edge detection method.

```
# Apply Canny edge detection
edges = cv2.Canny(gray_img, threshold1=30, threshold2=100)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.show()
```

In this step, we will convert the grayscale image to a binary image. This is done by setting a threshold value, and then all pixel values below the threshold are set to 0 (black), and all pixel values above the threshold are set to 255 (white).

```
# Set the threshold
threshold = 127

# Convert the grayscale image to a binary image
ret, binary_img = cv2.threshold(gray_img, threshold, 255, cv2.THRESH_BINARY)

# Display the binary image
plt.imshow(binary_img, cmap='gray')
plt.show()
```

In this step, we will display the grayscale image and the binary image side by side for comparison.

```
# Display the grayscale image and the binary image side by side
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(gray_img, cmap='gray')
axs[0].set_title('Grayscale Image')
axs[0].axis('off')

axs[1].imshow(binary_img, cmap='gray')
axs[1].set_title('Binary Image')
axs[1].axis('off')

plt.show()
```

In this step, we will replace the green channel of each pixel in the colored resized image with its blue channel.

```
# Create a copy of the resized image
modified_img = resized_img.copy()

# Replace the green channel with the blue channel
modified_img[:, :, 1] = modified_img[:, :, 0]

# Display the modified image
plt.imshow(cv2.cvtColor(modified_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```
cv2.imwrite('original_image.png', img)
cv2.imwrite('resized_image.png', resized_img_rgb)
cv2.imwrite('greyscale_image.png', gray_img)
cv2.imwrite('inverted_greyscale_image.png', neg_img)
cv2.imwrite('edge_detected_image.png', edges)
cv2.imwrite('binary_image.png', binary_img)
cv2.imwrite('modified_color_image.png', modified_img)
```

```
import numpy as np
# Combine all images side by side
combined_image = np.hstack(( resized_img_rgb, cv2.cvtColor(gray_img, cv2.COLOR_GRAY2BGR), cv2.cvtColor(neg_img, cv2.COLOR_GRAY2BGR), cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR), cv2.cvtColor(binary_img, cv2.COLOR_GRAY2BGR), modified_img))
# Display the combined image
plt.imshow(cv2.cvtColor(combined_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import numpy as np
# Combine all images in two rows
row1 = np.hstack((resized_img_rgb, cv2.cvtColor(gray_img, cv2.COLOR_GRAY2BGR), cv2.cvtColor(neg_img, cv2.COLOR_GRAY2BGR)))
row2 = np.hstack((cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR), cv2.cvtColor(binary_img, cv2.COLOR_GRAY2BGR), modified_img))
combined_image = np.vstack((row1, row2))
# Display the combined image
plt.imshow(cv2.cvtColor(combined_image, cv2.COLOR_BGR2RGB))
plt.show()
```
